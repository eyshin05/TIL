# 2.3 변동의 측정

```r
# 표준 편차는 일 별, 주 별, 월 별 판매량에 따라 어떻게 변할까?
```

```r
sales <- rnorm(365, mean=10, sd=3)
summary(sales)
```

```r
week_added_sales <- 0
month_added_sales <- 0

for ( week in 0:51 ) {
    week_added_sales[week+1] <- sum(sales[c( week*7+1:7 )])
}

for ( month in 0:11 ) {
    month_added_sales[month+1] <- sum(sales[c( month*30+1:30 )])
}
```

```r
summary(week_added_sales)
```

```r
summary(month_added_sales)
```

```r
c( "per day  : ", mean(sales), sd(sales), 
   "per week : ", mean(week_added_sales), sd(week_added_sales), 
   "per month: ", mean(month_added_sales), sd(month_added_sales) )
```

```r
c( sd(sales) / mean(sales) * 100 , 
   sd(week_added_sales) / mean(week_added_sales) * 100, 
   sd(month_added_sales) / mean(month_added_sales) * 100 )
```
