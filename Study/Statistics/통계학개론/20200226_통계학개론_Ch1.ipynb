# 연습문제 20.

```r
# 어느 회사의 남녀 별 입사 시험 성적
men_scores <- c(49, 86, 40, 45, 48, 93, 97, 58, 58, 98, 
                58, 82, 52, 56, 50, 85, 80, 60, 62, 80, 
                62, 72, 65, 60, 64, 70, 78, 67, 69, 88)
women_scores <- c(60, 72, 66, 65, 75, 78, 62, 64, 74, 58, 
                  68, 72, 67, 61, 62, 72, 79, 71, 74, 73)
```

```r
summary(men_scores)
```

```r
summary(women_scores)
```

```r
sqrt(30); sqrt(20)
```

```r
hist(men_scores, breaks=seq(30,100,by=10), xlim = c(0, 100))
hist(women_scores, breaks=seq(30,100,by=10), xlim = c(0, 100))
```

```r
stripchart(men_scores, xlim=c(0, 100))
stripchart(women_scores, xlim=c(0, 100))
```

```r
stem(men_scores)
stem(women_scores)
```
