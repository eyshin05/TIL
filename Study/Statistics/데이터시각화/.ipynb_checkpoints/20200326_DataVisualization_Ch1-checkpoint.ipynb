```r
options(repr.plot.height=4.5, repr.plot.width=7.5)

# type="n" 하면 실제로 점은 찍히지 않음
plot(c(0, 0), c(28, 60), type="n", xlim=c(0, 28), ylim=c(-5, 65),
     xlab="순서", ylab="재위기간", main="조선왕조",
     family="NanumGothic")

points(1:27, Period, pch=15, col="Red")

# x1, y1, x2, y2. lwd 는 line width. 
segments(1:27, rep(0, 27), 1:27, Period, lwd=3, col="Red")

abline(h=c(0, mean(Period)), lty="dotted", lwd=1, col="gray")
```

```r
options(repr.plot.height=4.5, repr.plot.width=7.5)

# type="n" 하면 실제로 점은 찍히지 않음
plot(c(0, 0), c(28, 60), type="n", xlim=c(0, 28), ylim=c(-5, 65),
     xlab="순서", ylab="재위기간", main="조선왕조",
     family="NanumGothic")

points(1:27, Period, pch=15, col="Red")

# segment 대체 코드
for (i in 1:27){
    lines(c(i, i), c(0, Period[i]), lwd=3, col="Red")
}

abline(h=c(0, mean(Period)), lty="dotted", lwd=1, col="gray")ㅣㅏㅏㅣㅏㅏㅏ
```

```r
options(repr.plot.height=8, repr.plot.width=7.5)

cum_period <- cumsum(Period)

plot(1:27, cum_period, type="n", 
     xlab="순서", ylab="누적연수", main="조선왕조",
     family="NanumGothic")

rect(0, 0, 1, cum_period[1], col="royalblue", border="royalblue")

for (i in 2:27){
    rect(i-1, cum_period[i-1], i, cum_period[i],
         col="royalblue", border="royalblue")
}

segments(0, 0, 27, 518, lty="dotted")
```

```r
plot(1:27, cum_period, type="n", 
     xlab="순서", ylab="누적연수", main="조선왕조",
     family="NanumGothic")

polygon(c(0, 0, 1, 1), 
        c(0, cum_period[1], cum_period[1], 0),
        col="royalblue", border="royalblue")

for (i in 2:27){
    polygon(c(i-1, i-1, i, i),
            c(cum_period[i-1], cum_period[i], cum_period[i], cum_period[i-1]),
            col=rainbow(27)[i], border="#00000000")
}

segments(0, 0, 27, 518, lty="dotted")
```
