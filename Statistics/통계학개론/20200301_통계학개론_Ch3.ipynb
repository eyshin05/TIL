```r
# 연습문제 13
x <- c(0:10)
y <- c(5, 8, 10, 12, 18, 14, 10, 9, 8, 4, 2)
prob <- y/sum(y)

df <- data.frame(x, y, prob)
df
```

```r
data <- c()

for (x in 1:11) {
    data = c(data, rep(x-1, y[x]))
}

hist(data, prob=T) -> h

h$counts <- cumsum(h$counts)
h$counts <- h$counts / max(h$counts)
plot(h)
```

```r
mean <- sum(df$x * df$prob)
mean
sum((df$x - mean)^2 * df$prob) # <-모표준편차
```

```r
sum((data - mean)^2) / 99 # <- 표본 표준편차
```
